```python
class Matrix:
    
    def __init__(self,size):
        board = []
        for i in range(size):
            tempArr = []
            for j in range(size):
                tempArr.append(False)
            board.append(tempArr)
        self.board = board
        
    def nextGeneration(self):
        x = nextGeneration(self)
        self.board = x.board
        
    def getNum(self):
        result = ""
        for i in range(size):
            for j in range(size):
                result += "1" if self.board[i][j] else "0"        

                
def getNum(matrix):
    result = ""
    for i in range(size):
        for j in range(size):
            result += "1" if matrix.board[i][j] else "0"       
    return result
# check if i,j inside the bounds of matrix
def inBounds(self,i,j):
    return i > 0 and i < len(self.board) and j > 0 and j < len(self.board)

# return if a cell live in the next generation
def amIAlive(self,i,j):
    countNeighbers = 0
    willLive = 3
    mayLive = 2
    # check each neighbor
    for a in range(i-1,i+2):
        for b in range(j-1,j+2):
            # don't count yourself
            if a==i and b==j:
                continue
            if inBounds(self,a,b) and self.board[a][b] == True:
                countNeighbers += 1
    if countNeighbers == willLive:
        return True
    if countNeighbers == mayLive and self.board[i][j] == True:
        return True
    return False

# compare two boards
def equals(matrix1,matrix2):
    m1 = getNum(matrix1)
    m2 = getNum(matrix2)
    return m1 == m2

# return a copy of a board
def copyBoard(self):
    m = Matrix(size)
    for i in range(len(self.board)):
        for j in range(len(self.board)):
            m.board[i][j] = self.board[i][j]
    return m

# calc how the matrix looks in the next generation
def nextGeneration(self):
    n = len(self.board)
    newMatrix = Matrix(n);
    for i in range(n):
        for j in range(n):
            newMatrix.board[i][j] = amIAlive(self,i,j)
    return newMatrix

# return if a configuration is new
def isNew(self,history):
    s = set()
    for step in history:
        if step.getNum() in s:
            return False
        s.add(step.getNum())
    return True
```

```python
# population

# update the population after one generation
def updatePopulation(pop,popByGrade):
    newPop = []
    for i in range(len(pop)):
        newPop.append(match(pop,popByGrade))
    return newPop

# create arr with all the population appear according to their grade
def popultionByGrade(population):
    r = []
    for i in population:
        if grade(i) == 0:
            r.append(i)
        for j in range(grade(i)):
            r.append(i)
    return r

# create initial population
def initializePopulation():
    population = []
    for i in range(numOfPopulation):
        m = setPopulation()
        population.append(m)
    return population

# returns a random matrix
def setPopulation():
    m = Matrix(size)
    numOfAliveCells = random.randint(minStart,maxStart)
    lim = 0
    while lim == 0:
        lim = random.randint(0,size - 1 - offset)
    while(numOfAliveCells > 0):
        row = random.randint(lim,lim+offset)
        col = random.randint(lim,lim+offset)
        if m.board[row][col] == False:
            m.board[row][col] = True
            numOfAliveCells -= 1        
    return m
```

```python
# evaluate solutions

# returns how many cells alive in a matrix
def howManyAlive(matrix):
    res = 0
    for i in range(size):
        for j in range(size):
             if matrix.board[i][j]:
                    res +=1
    return res

# get arr of matrixes and returns their grades as an arr
def lsOfGrades(pop):
    grades = []
    for i in pop:
        if grade(i) == 1:
            grades.append(0)
        else:
            grades.append(grade(i))
    return grades 

# calc biggest difference between the amount of alive cells in the best configuration in the next generation
# to the initial amout of alive cells
def grade(self):
    m = copyBoard(self)
    initialSize = howManyAlive(m)
    n = numOfGenerationUntilBiggest(m)
    for i in range(n):
        m = nextGeneration(m)
    return max(howManyAlive(m)-initialSize,1);


# returns in how many generations the configuration is the largest and the size of it
def numOfGenerationUntilBiggest(self):
    m = copyBoard(self)
    num = 0
    numOfGen = 0
    biggestSize = 0
    for i in range(numOfNewGenerations(m)):
        num = howManyAlive(m)
        if num > biggestSize:
            biggestSize = num
            numOfGen = i
        m = nextGeneration(m)
    return numOfGen

# calc how many generation takes until a configuration appears twice
def numOfNewGenerations(self):
    m = copyBoard(self)
    num = getNum(m)
    s = set()
    while True:
        if num in s:
            break;
        else:
            s.add(num)
        m = nextGeneration(m)
        num = getNum(m)
        #print(getNum(m) + "this is suppose to be a string represent matrix")
    return len(s)-1

# returns the size of a matrix in his last new configuration
def sizeInTheEnd(matrix):
    m = copyBoard(matrix)
    num = numOfNewGenerations(matrix)
    for i in range(num):
        m = nextGeneration(m)
    return howManyAlive(m)

# returns the size of a matrix in his last new configuration
def sizeInTheBiggest(matrix):
    m = copyBoard(matrix)
    num = numOfGenerationUntilBiggest(matrix)
    for i in range(num):
        m = nextGeneration(m)
    return howManyAlive(m)     
```

```python
# evolution

# chose parents from population by grades
def choseParent(population):
    flip = random.randint(0,len(population) - 1)
    return population[flip]

# choose two parents and return their son
def match(pop,popByGrade):
    
    # choose parents
    flip = random.randint(0,len(popByGrade)-1)
    dad = choseParent(popByGrade)
    mom = choseParent(popByGrade)
    while equals(dad,mom):
        dad = choseParent(population)
    # get birth
    return birth(mom,dad)

# return the son of two parents
def birth(mom,dad):
    son = Matrix(size)
    for i in range(size):
        for j in range(size):
            
            # recombination 
            whichParent = random.randint(0,1)
            if whichParent == 0:
                son.board[i][j] = mom.board[i][j]
            else:
                son.board[i][j] = dad.board[i][j]
                
            # mutation
            chance = random.randint(0,chanceToMutation)
            if chance == chanceToMutation:
                if son.board[i][j]:
                    son.board[i][j] = False
                else:
                    son.board[i][j] = True
    return son
```

```python
from tkinter import *
import math
import random

# variables
size = 10
chanceToMutation = 1000
numOfPopulation = 15
numOfGenertions = 30
cell_size = 20
minStart = 6
maxStart = 15
offset = 7
```

```python
# gui
def drawGrid(matrix):
    
    # intialize variables
    root = Tk()
    root.geometry("600x680")
    root.title("Game Of Life")
    my_canvas = Canvas(root,width = 600, height = 650,bg = "white")
    my_canvas.pack(pady = 0)
    next_button = Button(root, text="Next", command=lambda : run_and_canvas(my_canvas))
    next_button.pack()
    print("num of alive cells:" + str(howManyAlive(matrix)))
       
    emptyBoard(my_canvas)
    fillBoard(my_canvas,matrix)

    # btn func
    def run_and_canvas(my_canvas): 
        matrix.nextGeneration()
        emptyBoard(my_canvas)
        fillBoard(my_canvas,matrix)
        print("num of alive cells:" + str(howManyAlive(matrix)))
        
    root.mainloop()
    
def emptyBoard(my_canvas):
    for j in range(size):
        for k in range(size):
            my_canvas.create_rectangle(j*cell_size,k*cell_size,(j+1)*cell_size,(k+1)*cell_size,fill = "white")
            
def fillBoard(my_canvas,matrix): 
    for j in range(size):
        for k in range(size):
            if matrix.board[j][k]:
                my_canvas.create_rectangle(j*cell_size,k*cell_size,(j+1)*cell_size,(k+1)*cell_size,fill = "black")  
```

```python
# main

# set population

population = initializePopulation()
popByGrades = popultionByGrade(population)
biggest = 0

# initialize variables
bestMatrix = Matrix(size)
midBestMatrix = Matrix(size)
avgPerGen = []

for i in range(numOfGenertions+1):
    
    population = updatePopulation(population,popByGrades)
    popByGrades = popultionByGrade(population)
       
    # print grades for a generation each 5 generations
    if i % math.floor(numOfGenertions / 10) == 0:
        #print("hello2!")
        print("Generation: " + str(i))
        ls = lsOfGrades(population)
        print(ls)
        avgPerGen.append(sum(ls)/len(ls))
    j = 0
    
    # check each solution in the current population
    for p in population:
        num = grade(p)
        # update best solutions
        if j == math.floor(numOfGenertions / 2):
            midBestMatrix = bestBoard
            
        if num > biggest:
            biggest = num
            bestBoard = copyBoard(p)
        j += 1
        
print("-------------------------------------------------------------------------------")       
print("Biggest size - initial size: " + str(grade(bestBoard)))
print("Get biggest in " + str(numOfGenerationUntilBiggest(bestBoard))+ " generations")
print("Stabalize in " + str(numOfNewGenerations(bestBoard))+ " genertions")
print("Final size: " + str(sizeInTheEnd(bestBoard)))
print("Biggest size: " + str(sizeInTheBiggest(bestBoard)))
print(avgPerGen)
print("-------------------------------------------------------------------------------")   
drawGrid(bestBoard)
```
